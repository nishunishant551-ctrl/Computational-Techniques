```python
# Mean, Median and Mode for Grouped Data

# 1. Original Data
# xi → lower class limits of each class interval
# fi → upper class limits
# f → frequency of each class interval

xi = [15, 25, 35, 45, 55]    # Lower limits of classes
fi = [25, 35, 45, 55, 65]    # Upper limits of classes
f = [350, 375, 120, 80, 90]  # Frequencies of each class
x = []    # To store midpoints of each class
fx = []

# Displaying the original frequency table
print("Original Data:")
for i in range(len(f)):
    print(xi[i], "-", fi[i], ":", f[i])

# ---------------------------------------------------------------
# Calculate Midpoints and f*x
# Midpoint (x) = (Lower limit + Upper limit) / 2
# Then multiply each midpoint by its corresponding frequency
# ---------------------------------------------------------------
for i in range(len(f)):
    mid = (xi[i] + f[i]) / 2   # Calculate midpoint of each class
    x.append(mid)               # Add midpoint to list x
    fx.append(mid * f[i])       # Add (f*x) value to list fx

# Printing the midpoints
print("Midpoints:", x)

# ---------------------------------------------------------------
# Calculate Mean
# Formula: Mean = Σ(f*x) / Σf
# ---------------------------------------------------------------
mean = sum(fi) / sum(f)
print("\n1. Mean:", mean)

# ---------------------------------------------------------------
# Calculate Median
# Formula: Median = L + [(N/2 - cf_prev) / f_m] * h
# where:
# L = lower limit of median class
# N = total frequency (Σf)
# cf_prev = cumulative frequency before median class
# f_m = frequency of median class
# h = class width
# ---------------------------------------------------------------

cf = []      # Cumulative frequency list
temp = 0
for i in f:
    temp += i           # Add frequencies step by step
    cf.append(temp)     # Store cumulative frequency values

N = sum(f)              # Total frequency (Σf)
N2 = N / 2              # Half of total frequency

# Find the median class (where cf >= N/2)
for i in range(len(cf)):
    if cf[i] >= N2:
        medind = i      # Index of median class
        break

# Extract the required values for the median formula
L = xi[medind]                     # Lower limit of median class
h = fi[medind] - xi[medind]        # Class width
fmed = f[medind]                   # Frequency of median class
cfprev = 0 if medind == 0 else cf[medind - 1]  # CF before median class

# Apply median formula
med = L + ((N2 - cfprev) / fmed) * h
print("2. Median:", med)

# ---------------------------------------------------------------
# Calculate Mode
# Formula: Mode = L1 + [(f1 - f0) / (2f1 - f0 - f2)] * h
# where:
# L1 = lower limit of modal class
# f1 = frequency of modal class
# f0 = frequency before modal class
# f2 = frequency after modal class
# h = class width
# ---------------------------------------------------------------

max_f = max(f)                 # Highest frequency (modal class)
m = f.index(max(f))            # Index of modal class
L1 = xi[m]                     # Lower limit of modal class
f1 = f[m]                      # Frequency of modal class
f0 = 0 if m == 0 else f[m - 1] # Frequency before modal class
f2 = 0 if m == len(f) - 1 else f[m + 1] # Frequency after modal class
h = fi[m] - xi[m]              # Class width

# Apply mode formula
mode = L1 + ((f1 - f0) / ((2 * f1) - f0 - f2)) * h
print("3. Mode:", mode)
```
